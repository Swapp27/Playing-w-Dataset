```python
import seaborn as sb
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
sb.set_style(style = 'darkgrid')
```

```python
d = sb.load_dataset('tips')
```

```python
d.head()
```

```python
g = sb.FacetGrid(data = d,row= 'time',col = 'smoker',hue = 'sex')
g = g.map(sb.distplot,'tip',kde = False).add_legend()
#plt.tight_layout()
```

```python
sb.lmplot(data = d,x='total_bill',y = 'tip',row = 'sex',col = 'smoker',hue = 'time',markers = ['v','o'],palette = "magma")
#plt.tight_layout()
```
